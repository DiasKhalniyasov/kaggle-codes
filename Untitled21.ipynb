```python
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
from nltk.stem.wordnet import WordNetLemmatizer
import re
def cleaning(data):
    soup=BeautifulSoup(data).get_text()
    x= re.sub(r'\w*\d\w*', '', soup)
    x = str(x)
    for punct in "/-'":
        x = x.replace(punct, ' ')
    for punct in '&':
        x = x.replace(punct, f' {punct} ')
    for punct in '?!.,"#$%\'()*+-/:;<=>@[\\]^_`{|}~' + '“”’':
        x = x.replace(punct, '')
    x=re.sub('[^a-zA-Z]', ' ',soup)
    lower=str.lower(x).split()
    #a=[w for w in data.split()]
    words=set(stopwords.words('english'))
    a=[w for w in lower if not w in words]

    
    return a
```

```python
def clean_text(x):
    
    x = str(x)
    for punct in "/-'":
        x = x.replace(punct, ' ')
    for punct in '&':
        x = x.replace(punct, f' {punct} ')
    for punct in '?!.,"#$%\'()*+-/:;<=>@[\\]^_`{|}~' + '“”’':
        x = x.replace(punct, '')
    return x
def clean_numbers(x):

    x = re.sub('[0-9]{5,}', '#####', x)
    x = re.sub('[0-9]{4}', '####', x)
    x = re.sub('[0-9]{3}', '###', x)
    x = re.sub('[0-9]{2}', '##', x)
    return x
```

Building vocab to check cover %

```python
def build_vocab(sentences, verbose =  True):
    """
    :param sentences: list of list of words
    :return: dictionary of words and their count
    """
    vocab = {}
    for sentence in tqdm(sentences, disable = (not verbose)):
        for word in sentence:
            try:
                vocab[word] += 1
            except KeyError:
                vocab[word] = 1
    return vocab
```

```python
vocab = build_vocab(sentences)
```

```python
import operator 
def check_coverage(vocab,embeddings_index):
    b = {}
    oov = {}
    k = 0
    i = 0
    for word in tqdm(vocab):
        try:
            b[word] = embeddings_index[word]
            k += vocab[word]
        except:

            oov[word] = vocab[word]
            i += vocab[word]
            pass

    print('Found embeddings for {:.2%} of vocab'.format(len(b) / len(vocab)))
    print('Found embeddings for  {:.2%} of all text'.format(k / (k + i)))
    sorted_x = sorted(oov.items(), key=operator.itemgetter(1))[::-1]

    return sorted_x
```
